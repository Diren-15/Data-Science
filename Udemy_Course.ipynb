## *Google Playstore dataset Exploratory Data Analysis (EDA)*

```python
#importing important python files
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_store = pd.read_csv("googleplaystore.csv")
```

```python
df_store.head()
```

```python
df_store.info()
```

```python
#Generate descriptive statistics of numerical columns
df_store.describe()
```

```python
df_store.duplicated()
```

```python
#Handling Missing values
df_store.dropna(inplace = True) #replace missing values

# Dealing with duplicates
df_store.drop_duplicates(inplace = True) # remove duplicate rows
```

```python
df_store.info()
```

```python
# Change the data type of installs column to numerical data type and remove the comma and plus symbols
df_store['Installs'] = df_store['Installs'].str.replace('+','')  # Remove "+" sign
df_store['Installs'] = df_store['Installs'].str.replace(',','')  # Remove "," sign

df_store['Installs'] = df_store['Installs'].astype(int)  #Changing data type to int

#Change the data type of reviews colums to numerical data type
df_store['Reviews'] = df_store['Reviews'].astype(int)   #Converting data type to int
```

```python
df_store.head()
```

```python
# BAR PLOT
plt.figure(figsize = (10,5))
sns.countplot(data = df_store, x = 'Category')
plt.xlabel('Category')
plt.xticks(rotation = 90)
plt.ylabel('Count')
plt.title('Distribution of app categories')
```

```python
# Scatter Plot
plt.figure(figsize=(10,5))
sns.scatterplot(data = df_store, x = 'Reviews', y = 'Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.title('Relationship between Reviews and Rating')
```

```python
# Box Plot
plt.figure(figsize =(10,5))
sns.boxplot(data = df_store, x = 'Installs', y = 'Category')
plt.xlabel('Installs')
plt.ylabel('Category')
plt.title('Distribution of installs across categories')
```

```python
from scipy.stats import ttest_ind

# Splitting the dataset into free and paid apps
free_apps = df_store[df_store['Type'] == 'Free']
paid_apps = df_store[df_store['Type'] == 'Paid']
```

```python
# Performing t-test
t_stat, p_value = ttest_ind(free_apps ['Rating'], paid_apps ['Rating'])
```

```python
p_value
```

```python
#Checking the p_value
if p_value < 0.05:
  print('Reject Null Hypothesis: There is significant difference in ratings')
else:
  print('Fail to reject null hypothesis: There is no significant difference in ratings')
```

```python
#Scatter Plot of Ratings vs Installs
plt.figure(figsize = (10,5))
sns.scatterplot (data=df_store, x='Rating', y ='Installs', hue ='Category')
plt.xlabel('Ratings')
plt.ylabel('Number of Installs')
plt.title('Relationship between Rating and installs')
plt.legend(bbox_to_anchor=(1.05,1), loc =2, borderaxespad = 0.1)
plt.show()
```

```python
# Convert 'Last Updated' to Datetime
df_store['Last Updated'] = pd.to_datetime(df_store['Last Updated'])

# Set last Updated as the index
df_store.set_index ('Last Updated', inplace = True)

# Resample rating to monthly frequency
df_store_monthly = df_store['Rating'].resample('M').mean()
```

```python
# Line plot of average rating overtime
plt.figure(figsize=(10,5))
sns.lineplot(data = df_store_monthly)
plt.xlabel('Last Updated (Monthly)')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating trend overtime (Monthly)')
plt.show()
```
